```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 14


def calc_entropy(mu):
    return -mu*np.log2(mu) - (1-mu)*np.log2(1-mu)

mu_l = np.array([i/100 for i in range(1, 100)])
entropy = calc_entropy(mu_l)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(mu_l, entropy)
plt.ylabel("H[x]")
plt.xlabel("μ = p(x=1)")
plt.grid()
plt.show()
```

- エントロピーは「予測のしにくさ」
- μ=0.5 の時、x=0,1 のどちらの事象が起きるか予測しにくく、エントロピーは最大となる
