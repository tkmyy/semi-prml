```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 14
```

## 非線形変換の場合

```python
def p_x(x, mu, sigma): # gauss
    return (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))

mu = 1.2
sigma = 0.2

x = np.linspace(1e-2, 2, 100)

plt.plot(x, p_x(x, mu, sigma))
plt.title("p_x(x): gauss分布")
plt.grid()
plt.show()
```

```python
def p_y(y, mu, sigma): # y=exp(x)-1 (x=log(y+1))
    return p_x(np.log(y+1), mu, sigma)*(1/y)

y = np.exp(x) - 1

plt.plot(y, p_y(y, mu, sigma))
plt.title("p_y(y): 変数変換後の分布")
plt.grid()
plt.show()
```

```python
x_hat = x[p_x(x, mu, sigma).argmax()]
y_hat = y[p_y(y, mu, sigma).argmax()]


plt.figure(figsize=(8, 6))

plt.plot(x, p_x(x, mu, sigma), c="royalblue", label="p_x(x): N(1.2, 0.2)")
plt.plot(p_y(y, mu, sigma), y, c="limegreen", label="p_y(y)")
plt.plot(x, np.exp(x)-1, c="k", label="y=exp(x)-1\n( x=log(y+1) )")

plt.vlines(x_hat, ymin=0, ymax=np.exp(x_hat)-1, linestyles="--", colors="r")
plt.hlines(y_hat, xmin=0, xmax=np.log(y_hat+1), linestyles="--", colors="r")

plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.title("非線形変換後の最大値の位置の変化")
plt.show()
```

## 線形変換の場合

```python
def p_y(y, mu, sigma): # y=2x (x=y/2)
    return p_x(y/2, mu, sigma)*(1/2)

y = 2*x

plt.plot(y, p_y(y, mu, sigma))
plt.title("p_y(y): 変数変換後の分布")
plt.grid()
plt.show()
```

```python
x_hat = x[p_x(x, mu, sigma).argmax()]
y_hat = y[p_y(y, mu, sigma).argmax()]


plt.figure(figsize=(8, 6))

plt.plot(x, p_x(x, mu, sigma), c="royalblue", label="p_x(x): N(1.2, 0.2)")
plt.plot(p_y(y, mu, sigma), y, c="limegreen", label="p_y(y)")
plt.plot(x, y, c="k", label="y=2x\n( x=y/2 )")

plt.vlines(x_hat, ymin=0, ymax=2*x_hat, linestyles="--", colors="r")
plt.hlines(y_hat, xmin=0, xmax=y_hat/2, linestyles="--", colors="r")

plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.title("線形変換後の最大値の位置の変化")
plt.show()
```
