```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 14
```

```python
err = np.array([i/100 for i in range(-200, 201)]) # y - t
```

```python
def L_q(err, q):
    return abs(err) ** q
```

```python
for q in (0.01, 0.1, 0.3, 0.5, 1.0):
    L = L_q(err, q)
    plt.plot(err, L, label=f"q = {q}")

plt.xlabel("y - t")
plt.ylabel("|y - t| ** q")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid()
plt.show()
```

- q -> 0 の場合、y - t = 0 の時に0, それ以外は1となることがわかる
