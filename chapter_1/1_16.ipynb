```python
import math
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 16
```

```python
def calc_param_num(d, m):
    return math.factorial(d+m)/(math.factorial(d)*math.factorial(m))
```

```python
d_l = [i+1 for i in range(100)]
m_l = [i+1 for i in range(100)]

res = pd.DataFrame(index=d_l, columns=m_l)
for d in d_l:
    for m in m_l:
        res.loc[d, m] = calc_param_num(d, m)
```

```python
# M次のD次元多項式の総パラメータ数
# columns: 多項式の次数(M)
# index: 入力空間の次元(D)
res.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(res.index, res[3])

for d in (10, 100):
    plt.vlines(d, ymin=0, ymax=res[3][d], linestyles="--", colors="r")
    plt.hlines(res[3][d], xmin=0, xmax=d, linestyles="--", colors="r", label=f"D={d}: N(D, M)={res[3][d]}")

plt.xlabel("入力空間の次元: D")
plt.ylabel("総パラメータ数: N(D, M)")
plt.title("3次の多項式フィッティングにおけるパラメータ数の変化")
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.grid()
plt.show()
```
