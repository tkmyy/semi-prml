```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 14
plt.rcParams["axes.axisbelow"] = True
```

```python
plt.figure(figsize=(8, 4))

sample_num = 10000
for d in (2, 10, 100, 1000):

    # 平均0, 分散1とする
    mean = np.zeros(d)
    cov = np.eye(d)

    X = np.random.multivariate_normal(mean, cov, size=sample_num)

    # サンプリングしたデータ点のL2ノルムを計算
    norm_list = [np.linalg.norm(x, ord=2) for x in X]

    # rがどの辺りにデータが集中しているか
    mean_norm = np.mean(norm_list)
    plt.vlines(mean_norm, ymin=0, ymax=3500, linestyles="--", color="k")
    plt.hist(norm_list, alpha=0.5, label=f"{d}次元: r={np.round(mean_norm, 2)}")

plt.xlim(0, 40)
plt.title("多次元正規分布から生成されたデータの分布")
plt.xlabel("原点からデータ点までの距離(L2ノルム)")
plt.ylabel("データ数")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid()
plt.show()
```

→ 高次元になると分布の中心 (ここではr=0) にはほとんどデータが存在していない
